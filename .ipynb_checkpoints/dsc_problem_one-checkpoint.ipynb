# Data Science Challenge 2024 Task One

For this initial problem, we were required to identify normal and abnormal heartbeats using ptbdb_abnormal.csv and ptbdb_normal.csv found on:
https://www.kaggle.com/datasets/shayanfazeli/heartbeat/data



```python
%matplotlib inline
import pandas as pd
```

```python
abnormal_dataset_filepath = './dsc_onetwo/ptbdb_abnormal.csv'
normal_dataset_filepath = './dsc_onetwo/ptbdb_normal.csv'

abnormal_df = pd.read_csv(abnormal_dataset_filepath, header = None)
normal_df = pd.read_csv(normal_dataset_filepath, header = None)

abnormal_df
```

```python
normal_df
```

```python
full_df = pd.concat([normal_df, abnormal_df], axis=0)
X = full_df.iloc[:, :-1]
y = full_df.iloc[:, -1]
full_df.shape
```

# Visualizations:

# Model Training

There is a heavy class imbalance favoring abnormal heartbeats. We will need to accomodate for this during training

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from xgboost import XGBClassifier

X_train, X_test, y_train, y_test = tts(X, y, random_state=50, shuffle=True)

models = {
    'random forest': RandomForestClassifier(),
    'xgb': XGBClassifier()
}

accuracies = {}
f1s = {}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    accuracies[model_name] = accuracy
    
    
    f1 = f1_score(y_test, predictions)
    f1s[model_name] = f1

print(f"accuracies:{accuracies}\nf1 scores: {f1s}") 
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
cm = confusion_matrix(y_test, predictions)
print(f"confusion matrix:\n{cm}")
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 50)

X_res_train, y_res_train = sm.fit_resample(X_train, y_train)

model2 = XGBClassifier()
model2.fit(X_res_train, y_res_train)
predictions2 = model2.predict(X_test)
```

```python
accuracy2 = accuracy_score(y_test,predictions2)
f1_2 = f1_score(y_test, predictions2)
print(f"accuracy score: {accuracy2}\nf1 score: {f1_2}")
print(confusion_matrix(y_test, predictions2))
```

In this context, false negatives are worse since that would mean a patient with an abnormal heartbeat would be told it is normal.
Using SMOTE increased the amount of FNs so we will not be using SMOTE.

# Hyperparameter Tuning

```python
from scipy.stats import uniform, randint
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'learning_rate': uniform(0.01, 0.3),
    'n_estimators': randint(50, 200),
    'max_depth': randint(3, 10),
    'min_child_weight': randint(1, 6),
    'gamma': uniform(0, 0.5),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'reg_alpha': uniform(0, 0.1),
    'reg_lambda': uniform(0.8, 1.2)
}

random_search = RandomizedSearchCV(estimator=XGBClassifier(), param_distributions=param_dist, n_iter=50, scoring='accuracy', cv=3, verbose=1, random_state=42, n_jobs=-1)
random_search.fit(X_train, y_train)
print("Best Hyperparameters:", random_search.best_params_)
print("Best Accuracy:", random_search.best_score_)

```

```python
best_xgb_model = random_search.best_estimator_
best_xgb_model.fit(X_train, y_train)
y_pred = best_xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print(f"accuracy: {accuracy}\nf1: {f1}")
```
